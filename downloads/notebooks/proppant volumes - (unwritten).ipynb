### 4.3. Factors that affect EUR: Completion size and lateral length

The trend observed in <a href="#fig-6">Figure 6</a> above, where EURs increase with increasing DUC time, contradicts the established scientific knowledge. However, we have found that this illogical trend is due to the use of smaller completion sizes and shorter lateral lengths for wells with shorter DUC times. Interestingly, the basins showing the greatest difference in EUR between the peak and trough in <a href="#table-1">Table 1</a> (Anandarko and Haynesville) are also those with the largest increase in completion size and lateral length as DUC time increases.

<p></p>
<p style="display: inline"><a href="#fig-7">Figure 7</a> illustrates the variation in EUR due to completion size and lateral length with increasing DUC time for different basins. The largest increase in EUR between two time periods ([1] and [2]) was observed in the Haynesville basin, with a difference of 2,223 MBOE (a 209% increase)(<a href="#table-2">Table 2</a>)</p><strong style="color: #a4a4a4">(<img src="jupyter_images/duc_legend_1.png" class="inline-image"><p style="display: inline">grey bars)</strong>. A correlation analysis showed that this increase was positively associated with a 46% increase (2,111 ft) in lateral length (<a href="#table-3">Table 3</a>)<strong style="color: #111111">(<img src="jupyter_images/duc_legend_2.png" class="inline-image">black line)</strong>, 121% boost in PPF (<a href="#table-3">Table 3</a>)</p><strong style="color: #9504d3">(<img src="jupyter_images/duc_legend_3.png" class="inline-image"><p style="display: inline">purple lines)</strong> and a 95% boost in FPF (<a href="#table-4">Table 4</a>)</p><strong style="color: #287cb7">(<img src="jupyter_images/duc_legend_4.png" class="inline-image"><p style="display: inline">blue lines)</p></strong><p style="display: inline"> for wells completed between 200-250 days after drilling, compared to wells completed within 50 days after drilling. Similar trends were observed for the Anadarko, Permian, and Bakken basins, although the EUR increase in the Bakken basin occurred at a much slower pace over a 500-day period.</p>

Recall that PPF and FPF represent completion size normalized by lateral length (per foot). Increase in PPF and FPF demonstrates that not only the operators are increasing the absolute volume of proppant and fluids to fracture the additional lateral length, but also increasing the completion concentration per each foot of the lateral for the older DUC wells. It can be speculated that the operators are applying bigger completion size per lateral length to make up for the potential reservoir depletion problems that the older DUC wells may suffer from due to nearby parent wells. However, the drastic 217% (Andarko) and 209% (Haynesville) jump in EUR (<a href="#table-2">Table 2</a>)<strong style="color: #a4a4a4">(<img src="jupyter_images/duc_legend_1.png" class="inline-image">grey bars)</strong> with boosted completion concentration between [1] and [2] shows that the <strong>operators tend to undersize their completion design for wells completed shortly after drilled (up to 250 days), resulting in sub-optimal EUR.</strong> In the other words, wells completed shortly after drilled tended to be completed with smaller completion concentration. The operators may profit more for larger completion investment for wells completed right after drilled.

In contrast, wells in the Appalachia, Eagle Ford, and Niobrara basins showed only a slight increase in completion size and lateral length with increasing DUC time, with the exception of the Niobrara basin which showed a decrease in PPF (<a href="#table-4">Table 4</a>)<strong style="color: #9504d3">(<img src="jupyter_images/duc_legend_3.png" class="inline-image">purple line)</strong> despite the longer lateral length. The magnitude of the difference between [1] and [2] was not as drastic as in the other basins. This finding explains why wells in these basins showed minimal difference in EUR with increasing DUC time in contrast to the Anadarko, Bakken, Haynesville, and Permian basins; because the volume of proppants and fluids used for completion and lateral length stayed relatively constant with increasing DUC time. Further data processing is necessary to remove the impact of completion size and lateral length and observe the sole influence of DUC time on EUR. 

Additionally, it is worth noting that the wells with longer lateral length tended to have longer DUC time. While this may sound obvious, taking a closer look at it reveals that it is not so. Between [1] and [2], the average difference in lateral length of wells in the Haynesville basin was 2,111 ft and the difference in time was 200 days. An experienced drilling and completion engineer would recognize that it does not take an additional 200 days to drill an extra 2,111 ft and hydraulically fracture that additional length. Modern day drilling takes under 15~25 days, and completion 3~5 days. One possible explanation for operators' tendency to delay completion for longer lateral wells is that the costs associated with bigger proppant and fluid volumes required to adequately fracture longer lateral sections are higher. Nevertheless, further research is needed to fully understand the economic trade-offs involved in completion timing for longer lateral wells.
