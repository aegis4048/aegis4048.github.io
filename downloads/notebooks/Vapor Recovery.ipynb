### Example 1. VRT vapor recovery

The above discussions on VFD/PID loop (<a class="internal-link" href="#8.1.-VFD-and-PID-Loop">Section 8.1</a>) and recycle mode (<a class="internal-link" href="#8.2.-Recycle-mode">Section 8.2</a>) used VRT vapor recovery scenario as an example. VRTs typically require 2-4 psig operating range. Above 4 psig risks a chance of gas blowby to tanks, while below 2 psig risks a chance of liquid overflow to gas line, and into a VRU suction scrubber, triggering high liquid level shutdown on a VRU. This is due to the unique design of a VRT, which I do not cover in detail here.

<a class="internal-link" href="#fig-45">Figure 45</a> and <a class="internal-link" href="#fig-46">46</a> below shows a simplified layout and P&ID of a VRT vapor recovery system. Assuming 5 psig is the maximum pressure you want on a VRT (5 psig is for liquid level control, VRTs typically have 75-150 psig MAWP rating), there will be a BPV set at 5 psig for overpressure relief to combustor during mechanical failure of a compressor or volume surge (see <a class="internal-link" href="#note_3">Note 3</a> for combustor relief working principles). The VRU targets to maintain the VRT pressure between 2-4 psig by ramping up/down a VFD. Gas flows preferentially to the VRU side due to its lower suction pressure (2-4 psig) compared to the flare side BPV's 5 psig resistance, making the VRU the path of least resistance.

    <li><u>3 psig:</u> Target setpoint</li>
    <li><u>2 psig:</u> Recycle setpoint</li>
    <li><u>Between 1-2 psig:</u> Recycle operating range</li>

### Example 2. Tank vapor recovery

Different pressure control applications (tanks, heaters, separators, wellheads) require different operating pressures. The most challenging is vapor recovery from atmospheric storage tanks at upstream production sites due to their extremely low operating range (measured in ounces, 1-16 oz). The tank vapor recovery has six notable components: 1) <u>API 12F tank design</u>, 2) <u>Tank vent header</u>, 3) <u>Thief hatch</u>, 4) <u>Enardo valve</u>, 5) <u>Combustor</u>, and 6) <u>Pressure transmitter</u>.


<div class="centered-title">
<span class="title-text mt-2">6 Major Components of Tank Vapor Recovery</span>
</div>

<ol class="numbered-bullets mt-4">
<li style="margin-bottom: 1.5em !important;" id="api_12f_tank_design">
<p><strong>API 12F tank design</strong></p>
<p>API 12F specifies standardized atmospheric storage tanks for upstream oil and gas operations. The 2008 12th edition recommends 8-16 oz MAWP tanks with 4" vent headers, while the 2019 13th edition upgrades to 16 oz MAWP with 6" headers - providing greater operating margin and reduced friction loss for vapor recovery and combustion systems.</p>
</li>
<li style="margin-bottom: 1.5em !important;" id="tank_vent_header">
<p><strong>Tank vent header</strong></p>
<p>Vent headers are oversized (4-6" per API 12F) to minimize friction loss - notably larger than or equal to separator gas lines handling 100x more volume. This is critical because while separators operate at 50-300 psig, tanks only maintain 0.1-1 psig (1-16 oz). Preserving this minimal pressure in ounces is essential for proper combustor operation.</p>
</li>
<li style="margin-bottom: 1.5em !important;" id="thief_hatch">
<p><strong>Thief hatch</strong></p>
<p>These 8" pressure-vacuum safety hatches serve as last-resort relief devices. They are notorious for emission issues. Their major flaw: once opened, they remain open until manually reset (requiring physical force to "click" the latch shut). This creates prolonged emissions between activation and manual closure, which is why they should always be "last-resort", secondary pressure relief to PRVs set at slightly lower pressures than thief hatches.</p>
</li>
<li style="margin-bottom: 1.5em !important;" id="enardo_valve">
<p><strong>Enardo valve</strong></p>
<p>A specialized PRV for low-pressure (ounce-range) applications featuring stackable weighted pallets that allow adjustable backpressure. These serve dual purposes: as primary overpressure protection (set below thief hatch activation) and as backpressure regulators. The pallet weights exert backpressure (ex: 14 oz for tank applications) to preferentially drive vapors to the VRU suction line, which maintains lower pressure (ex: 4-13 oz) due to the VRU's vacuum - the higher resistance path through the Enardo/combustor line ensures flow follows this path (VRU suction) of least resistance when the VRU is online. During process upsets or VRU shutdown, the valve relieves excess pressure through either: 1) Vent-to-atmosphere configuration (used in VRT applications as shown in <a class="internal-link" href="#fig-45">Figure 45</a> and <a class="internal-link" href="#fig-46">46</a> where the VRU inlet is occupied by the VRT gas line), or 2) Pipe-away configuration (for tank vapor recovery systems shown in <a class="internal-link" href="#fig-47">Figure 47</a> and <a class="internal-link" href="#fig-48">48</a>). Tank vapor recovery uses pipe-away configuration that dumps to combustor during high pressure events (see <a class="internal-link" href="#note_3">Note 3</a> for combustor relief working principles).</p>
</li>
<li style="margin-bottom: 1.5em !important;" id="combustor">
<p><strong>Combustor</strong></p>
<p>Combustors provide VOC destruction when VRUs are offline or overloaded. These units require minimum "turn on" pressures (ex: 1-4 oz) to initiate operation. Like VRUs, their capacity increases with inlet pressure - hence the need for oversized vent headers to minimize pressure drop. Insufficient pressure reduces capacity, potentially forcing direct venting to atmosphere through PRVs or thief hatches.</p>
</li>
<li style="margin-bottom: 1.5em !important;" id="pressure_transmitter">
<p><strong>Pressure transmitter</strong></p>
<p>A pressure transmitter (PT) must be installed either directly on the tank roof port or within the roof vent manifold piping, with direct wiring to the compressor control panel. Ground-level installation is to be avoided as it produces false low readings from friction losses - an unacceptable error given the system's extremely low operating pressures in ounces. See <a class="internal-link" href="#note_2">Note 2</a> for for more details. The PT sends pressure feedback to compressor PLC to control tank pressure.</p>
</li>
</ol>

<div class="alert alert-info" id="note_2">
    <h4>Note 2: PT placement requirement</h4>
    <p>Pressure transmitters (PTs) must be installed as close as possible to source vessels to ensure faster and more reliable pressure control. While most compressors include internal PTs on suction lines, these are prone to false readings from pipeline pressure drops between the vessel and suction scrubber inlet, as well as response delays from gas travel time.</p>
    <p>Heater treater and VRT applications operating in pounds can tolerate some measurement inaccuracies due to their wider operating margins. However, <strong>tank applications require PTs to be mounted directly on either the tank roof or tank vent manifold</strong>, with direct wiring to the compressor control panel. Ground-level installations will register false low pressures due to friction loss. For instance, in the above <a class="internal-link" href="#8oz_tank">8 oz MAWP tank scenario</a>, a ground-level suction line PT reading 6 oz (with actual tank pressure at 7-8 oz due to 1-2 oz pipeline losses) could prematurely trigger the 7 oz PRV to flare or 8 oz thief hatch to vent.</p>
    <p>Response delays compound these issues. When tank pressure rises from flash gas events, delayed VFD response means the compressor won't immediately ramp up to process the increased volume. This creates a temporary backlog in the suction line, increasing backpressure on the tank vent header and potentially causing venting/flaring before the compressor can respond.</p>
    <p>Low-pressure scenarios present similar risks. Without gas/N2 blanketing systems (uncommon at upstream sites), false PT readings may cause the compressor to overdraw tanks, activating vacuum safety valves. These valves introduce oxygen into the system to prevent tank implosion, creating midstream pipeline corrosion risks from air ingress.</p>
</div>

    <li><u>Above 50 psig:</u> Compressor high suction pressure shutdown</li>
    <li><u>Betweem 32-50 psig</u>: Operating margin</li>
    <li><u>Betweem 28-32 psig:</u> Normal operating range</li>
    <li><u>30 psig:</u> Target setpoint</li>
    <li><u>28 psig:</u> Recycle setpoint</li>
    <li><u>Between 20-28 psig:</u> Operating margin</li>
    <li><u>Below 20 psig:</u> Low pressure shutdown (flexible as long as above 20 psig)</li>
</ol>
</div>

Similar to VRT and tank applications, the gas line splits into two paths: 1) to VRU suction (main path), and 2) to flare with a back pressure valve (BPV) set higher than the target suction pressure and lower than heater MAWP. This pressure differential ensures vapors flow to the VRU during normal operation (path of least resistance), while automatically diverting to flare for pressure relief during VRU shutdowns or process upsets. See <a class="internal-link" href="#fig-A">Figure A</a> below for an example telemetry data showing pressure relief to flare during VRU shutdown. See <a class="internal-link" href="#fig-49">Figure 49</a>, <a class="internal-link" href="#fig-50">50</a>, and <a class="internal-link" href="#fig-51">51</a> for example P&IDs for BPV (BPV-6001) installations on flare side.

<hr>

<p>Heater vapor recovery differs from VRT/tank applications in <strong><u>potentially requiring a suction control valve (SCV)</u></strong> (see <a class="internal-link" href="#7.9.-Suction-control-valve">Section 7.9</a> for SCV working principles). Given heaters typically operate at 20-70 psig and compressors have 50 psig suction limits:</p>

<div class="ordered-list ordered-list-note ordered-list-nested">
<ol>
    <li><strong>Above 50 psig:</strong> SCV mandatory to reduce pressure below compressor suction limit</li>
    <li><strong>40-50 psig:</strong> SCV "highly recommended" due to inevitable occasional pressure swings</li>
    <li class="bottom-border"><strong>20-40 psig:</strong> SCV optional but recommended for upstream applications where:</li>
        <div class="nested-li-wrapper ml-40">
            <li>Separators upstream of a heater may not fully absorb flow variations (typical 2x-6x normal swings)</li>
            <li>Compressor response may lag behind sudden gas surges</li>
            <li>Historic data shows severe pressure fluctuations</li>
        </div>
    <li value="4"><strong>40-50 psig:</strong> SCV "highly recommended" due to inevitable occasional pressure swings</li>
</ol>
</div>

<p>While compressors can adjust RPM to maintain target pressure, sudden surges often exceed this capacity. Operators should review heater telemetry history - if uncertain, the <span>&#36;</span>1,500-<span>&#36;</span>2,500 SCV installation cost is typically justified versus frequent shutdown risks. Retrofit remains an option if initial SCV-less operation proves problematic.</p>

<p>SCV installation also helps maintain VRU operation even during flare activation. When flare engages (indicating system pressure exceeded BPV setpoint), the SCV prevents compressor shutdown by isolating suction pressure from system (heater vessel + heater gas line piping) pressure. <a class="internal-link" href="#fig-49">Figure 49</a> shows a heater treater system without SCV (25 psig target, 110 psig BPV) where flow surges may cause complete VRU shutdown - system pressure, including the compressor suction pipe, equalizes up to BPV setpoint, triggering high HP/suction pressure shutdowns and 100% flare diversion. 
    
<p><a class="internal-link" href="#fig-50">Figure 50</a> demonstrates the SCV-equipped system where the valve physically isolates compressor suction pressure from upstream system pressure. During surges, the SCV maintains suction pressure at ≤25 psig (regardless of heater/BPV side pressure) by throttling flow, while allowing only excess volume beyond compressor capacity to reach the 110 psig BPV setpoint and divert to flare. This enables continuous VRU operation handling most surge volume (e.g., 70%), with only the remainder (e.g., 30%) flaring - a critical improvement over the non-SCV configuration's complete shutdown.</p>

<p><a class="internal-link" href="#fig-51">Figure 51</a> demonstrates a specialized configuration where operators maintain heater pressure at 70 psig (above the compressor's 50 psig suction limit) using two pressure controls: (1) a 70 psig BPV on the suction line, and (2) an SCV set below the limit (e.g., 25 psig). While uncommon, this setup serves specific cases where high heater pressure must be preserved alongside VRU operation, creating three distinct pressure zones: heater side (70 psig), SCV-regulated compressor suction (25 psig), and flare protection (110 psig).</p>

### Example 4. Separator drawdown

<p>The term "drawdown" refers to artificially reducing wellhead backpressure by lowering salesline pressure. When wellhead pressure decreases, the well responds with increased flowrates. Between wellhead and salesline (150 psig in this example), a primary separator processes the flow: gas moves to salesline while liquids (oil/water) proceed to heaters then atmospheric storage tanks. In <a class="internal-link" href="#fig-52">Figure 52</a>, check valve CHK-6000 acts as a backpressure valve, remaining closed until separator pressure reaches salesline pressure. This accumulated separator pressure creates backpressure that propagates through the wellhead tubing flowline to the downhole BHA and reservoir - higher surface backpressure increases backpressure on downhole reservoir and reduce flowrates.</p>

<p>Installing a compressor between separator and salesline alters this dynamic. The compressor isolates the wellhead from salesline pressure, exposing it only to the compressor's suction pressure (adjustable between 0-50 psig) instead of the full 150 psig salesline pressure plus friction losses. By consuming energy (electricity/fuel gas), the compressor boosts pressure to meet salesline requirements while enabling significant pressure drawdown at the wellhead. During normal operation, all separator gas flows to the compressor suction side (path of the lowest resistance) and the wellhead sees minimal back pressure. During compressor shutdown or volume surge, the separator pressure builds upto the salesline pressure, exerting high backpressure on the wellhead.</p>

The below are five key considerations for separator drawdown with screw compressors:

<div class="centered-title">
<span class="title-text mt-2">Operation Considerations</span>
</div>

<ol class="numbered-bullets mt-4">
<li id="application_type" style="margin-bottom: 1.5em !important;">
<p><strong>Application Types</strong></p>
<p>Separator drawdown applications include:</p>
<ul>
<li><p><u>Flowing wells:</u> Naturally producing wells where reduced wellhead pressure increases flowrates without requiring artificial lift.</p></li>
<li><p><u>Liquid-loaded wells:</u> Surface pressure reduction decreases the critical gas flowrate needed for downhole liquid removal through the vertical section of a production tubing, eliminating the need for well shutdowns to build-up downhole pressure, or swabbing operations to mechanically unload liquids. Note that these shutdowns or swabbings need to be done a regular basis because the liquids will accumulate again.</p></li>
<li><p><u>Plunger-lift wells:</u> Used both when plunger lift alone is insufficient to remove liquid due to low gas volume <strong>or</strong> for multi-well pads where some wells struggle with liquid loading while others produce adequately. Even for wells producing fine with plungers benefit from reduced bottomhole pressure, responding with production uplift. One compressor can service multiple wells by synchronizing plunger cycles and using compressor auto-start features (triggers upon detecting pressure on a suction line, by a pressure transmitter feedback to the PLC), making this the most cost-effective configuration for multi-well pads.</p></li>
<li><p><u>Replacing gas-lift wells:</u> Separator drawdown offers an alternative for wells lacking pipeline infrastructure for buyback gas or facing expensive buyback sources. When produced gas flowrates are insufficient for gas lift and buyback options are unavailable or costly, drawdown compression can provide a more economical solution. Economic feasibility should be verified beforehand.</p></li>
</ul>
<p>All applications require production gains to justify compressor rental costs (anywhere between <span>&#36;</span>2,000-<span>&#36;</span>7,000/month for screw compressors, depending on capacity).</p>
</li>
<li id="suction_control_valve" style="margin-bottom: 1.5em !important;">
<p><strong>Suction Control Valve</strong></p>
<p>An SCV (see <a class="internal-link" href="#7.9.-Suction-control-valve">Section 7.9</a> for SCV working principles) is mandatory as separators experience constant 2x-6x pressure/flowrate swings directly from wellheads. Without SCV protection, compressors suffer frequent high-HP/suction pressure shutdowns.</p>
</li>
<li id="pressure_targets" style="margin-bottom: 1.5em !important;">
<p><strong>Pressure Targets</strong></p>
<p>While compressors can pull separator pressure to vacuum, suction is typically maintained at 25-45 psig because: (1) flowrate capacity decreases at lower pressures (see <a class="internal-link" href="#table-1">Table 1</a>), often offsetting production gains by setting it at vacuum; and (2) separators require minimum 20 psig to reliably dump liquids to downstream vessels without a liquid pump installation (increased cost/maintenance). The pressure target is maintained by a skid-contained PT on a suction scrubber sending pressure feedback to a compressor control panel.</p>
</li>
<li id="non_invasive_surface_installation" style="margin-bottom: 1.5em !important;">
<p><strong>Non-Invasive Surface Installation</strong></p>
<p>The entire installation occurs at surface level downstream of the primary separator. Setup requires only hose connections for compressor inlet/outlet and a few additional control valves. The system is completely reversible - if production uplift doesn't justify operation, the compressor can be removed and piping restored to its original configuration with minimal effort.</p>
</li>
<li id="salesline_pressure" style="margin-bottom: 1.5em !important;">
<p><strong>Salesline Pressure</strong></p>
<p>Higher salesline pressures offer greater production uplift potential due to larger achievable pressure drops after compressor installation. Wells with salesline pressures above 100 psig are good candidates for testing, while those above 150-200 psig are strongly recommended for trial.</p></li>
</ol>

### Example 5. Casing drawdown

<blockquote>
  <p><strong>BEFORE YOU READ:</strong> I recommend you read the above <a class="internal-link" href="#Example-4.-Separator-drawdown">Example 4: Separator Drawdown</a> before reading this section to help with your understanding, as these two are similar.</p>
</blockquote>

The purpose of casing drawdown is to eliminate gas interference in downhole pumps like sucker rod pumps (SRPs) and electric submersible pumps (ESPs). Gas interference reduces pump efficiency, damages equipment, and forces frequent shutdowns for repairs.  

Traditional fixes—such as installing downhole separators, optimizing pump placement, or increasing wellhead back pressure—are limited because gas must still travel up the tubing alongside liquids. Casing drawdown addresses the root cause by opening the casing side, creating a dedicated gas relief path.  

Consider a wellhead with 150 psig sales line backpressure at the separator. The tubing flowline delivers fluids to the separator for gas/liquid separation, with liquids routed to heater treaters and gas to salesline. As reservoir pressure declines, it can no longer lift liquids up the tubing unaided. An SRP (ESP works too) is installed but suffers from gas interference.  

<hr>

<strong><u>The solution:</u></strong>

Connect the casing side to a compressor inlet set at 5 psig suction pressure (adjustable 0-50 psig). The tubing remains at 150 psig salesline backpressure while the casing sees just 5 psig. This 145 psig differential (150-5) drives gas through the casing (the path of least resistance), while liquids are restricted to the tubing because: (1) depleted reservoir pressure lacks energy to push liquids up naturally, and (2) the casing's larger cross-section requires greater lift pressure than the tubing. Recall that there's no pump on the casing side.

<strong><u>Result</u></strong>

The wellbore practically functions as a two-phase separator. Gas exits through the casing (to compressor) and liquids through the tubing (via SRP), eliminating gas interference. The compressor suction doesn't pull liquid due to the absence of SRP on the casing side, while the SRP maintains steady liquid production unaffected by gas.

<hr>

While fixing gas interference in downhole pumps is the main purpose of casing drawdown, it also increases gas flow rates. Unlike conventional systems where gas harms pumps, here it flows exclusively up the casing. Reducing casing backpressure from 150 psig to 5 psig typically boosts gas flow rates 2x-6x. This is similar to increasing flowrates by lowering salesline backpressure, explained in <a class="internal-link" href="#Example-4.-Separator-drawdown">Example 4: Separator Drawdown</a>.

The uplift is greater for wells intentionally holding higher wellhead backpressure than salesline pressure. This achieves two things: 1) pressure increase shifts the operating point on the phase envelope, reducing bubble-out volume, and 2) compresses gas, decreasing absolute volume and improving pump fillage %. For SRPs, pressures beyond 200 psig show diminishing returns.

Assume an operator maintains 200 psig wellhead pressure when salesline pressure is 120 psig. Installing casing drawdown creates two distinct pressure regimes:

- <p class="mt-0">Tubing side: 120 psig salesline backpressure (200-120 = 80 psig artificial backpressure removed)</p>
- <p class="mt-0">Casing side: 5 psig compressor suction pressure</p>

Liquid flowrates increase due to removal of 80 psig artifical backpressure on tubing. Gas flowrate increases on the casing side due to 195 psig (200-5) backpressure reduction, which can easily come back with 4x-6x gas rate increase. 

The below are five key considerations for casing drawdown with screw compressors:

<div class="centered-title">
<span class="title-text mt-2">Operation Considerations</span>
</div>

<ol class="numbered-bullets mt-4">
<li id="test_separator" style="margin-bottom: 1.5em !important;">
<p><strong>Test Separator</strong></p>
<p>While casing drawdown keeps most liquids flowing up the tubing, some may flow up the casing due to well slugging behavior. Wells experience frequent pressure and flow rate swings that may temporarily exceed the critical gas flow rate needed to lift liquids up the casing annulus. Entrained liquid in mist form may also travel with gas. A screw compressor's skid-mounted suction scrubber isn't designed as a separator - it only handles small amounts of entrained liquid in gas lines that have already been separated.</p>
<p style="margin-top: 1rem !important;">The liquid volume may be significant enough to trigger high-level shutdowns in the suction scrubber (see <a class="internal-link" href="#fig-33">Figure 33</a>). Predicting liquid flowrate up the casing is difficult as these come from entrained mist or liquid slugs from occasional pressure swings that are difficult to be modeled and predicted. An operator may either try directly hooking up the casing with compressor inlet, and see if a compressor experiences frequent high liquid level shutdowns. If the problem persists, a trailer-mounted test separator, or a small permanent external scrubber may be installed to handle liquids. Or they can just begin with a test separator, observe how the well responds, and then determine whether a permanent scrubber installation is needed.</p>
</li>
<li id="suction_pressure_setpoint" style="margin-bottom: 1.5em !important;">
<p><strong>Suction Pressure Setpoint</strong></p>
<p>Suction pressure can be lowered to vacuum, but 5 psig is typically recommended. This balances lowering casing pressure while preserving compressor flow rate capacity (see <a class="internal-link" href="#table-1">Table 1</a>). When using an external separator (e.g., trailer-mounted test separator), 5 psig pressure is insufficient to dump liquid to downstream vessels, requiring a liquid pump (see <a class="internal-link" href="#fig-53">Figure 53</a> and <a class="internal-link" href="#fig-54">54</a>). If installing a liquid pump is undesirable, operators can maintain 20-30 psig suction. However, the production uplift from maintaining 5 psig suction usually outweighs pump operating costs, as casing-side liquid volumes are typically small.</p>
</li>
<li id="capacity_sizing" style="margin-bottom: 1.5em !important;">
<p><strong>Capacity Sizing</strong></p>
<p>Gas flowrate will increase after compression. Size the units for 2-3x current flowrates if above 200 MSCFD, 3x if above 100 MCFD, and up to 4x if under 100 MCFD. Note that these are ballpark numbers based on experience.</p>
</li>
<li id="application_type" style="margin-bottom: 1.5em !important;">
<p><strong>Application Types</strong></p>
<p>Casing drawdown applications include:</p>
<ul>
<li><p><u>SRP wells:</u> Casing pressure can be set anywhere between 5-50 psig depending on operational needs. The casing pressure is maintained by a skid-contained pressure transmitter PT-5000 on a suction scrubber sending feedback to a control panel PLC-5000 (see <a class="internal-link" href="#fig-53">Figure 53</a>). The pressure control mechanisms with PT-PLC feedback are similar to those of <a class="internal-link" href="#Example-3.-Heater-vapor-recovery">Heater vapor recovery</a> and <a class="internal-link" href="#Example-4.-Separator-drawdown">Separator drawdown</a> applications.</p></li>
<li><p><u>ESP wells:</u> ESPs need 70-90 psig on the casing. This is above the 50 psig max suction limit of a compressor, requiring mandatory SCV installation between casing and compressor inlet. A pressure transmitter PT-1000 on the casing sends pressure feedback to control panel PLC-5000 and maintains desired 70-90 psig on the casing (see <a class="internal-link" href="#fig-54">Figure 54</a>). The pressure control mechanisms with PT-PLC feedback are similar to those of <a class="internal-link" href="#Example-2.-Tank-vapor-recovery">Tank vapor recovery</a> in a sense that the PT is directly mounted on the pressure source (casing vs. tank) to control its pressure. The key difference is that while the tank pressure is within the compressor suction limit, but the ESP casing pressure is beyond the compressor suction limit.</p></li>
</ul>
<p>Both are types of downhole pumps that suffer from gas interference problems.</p>
</li>
<li id="suction_control_valve" style="margin-bottom: 1.5em !important;">
<p><strong>Suction Control Valve</strong></p>
<p>An SCV (see <a class="internal-link" href="#7.9.-Suction-control-valve">Section 7.9</a> for SCV working principles) is mandatory as wellheads frequently experience 2x-6x pressure/flowrate swings directly from wellheads. Without SCV protection, compressors suffer frequent high-HP/suction pressure shutdowns.</p>
</li>
</ol>
