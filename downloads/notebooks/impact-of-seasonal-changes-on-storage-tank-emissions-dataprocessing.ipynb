```python
import pandas as pd
import numpy as np

name = 'Worksheet in ProMaxBook1.xlsm'
name = 'Worksheet in ProMaxBook1.xlsm'

df = pd.read_excel(name, sheet_name='Sheet2')

df = pd.read_excel('seasonal_impact_data.xlsx', sheet_name='Sim Format 1')
cols = list(df.columns[1:])
df
```

```python
dfs = []
for col in cols:
    df_temp = df.groupby('Location')[col].apply(lambda df: df.reset_index(drop=True)).unstack().reset_index()
    df_temp.insert(1, 'Results', col)
    dfs.append(df_temp)
    
df_final = pd.concat(dfs, axis=0).sort_values(by=['Location', 'Results']).reset_index(drop=True)
df_final
```

```python
# Custom sort order for the "Results" column
custom_sort_order = ['AVG Air T (F)', 'Gas (MCFD)', 'Tank T (F)', 'RVP (psia)', 'SG', 'Dew T (F)', 'GHV (Btu/scf)']
df_final['Results'] = pd.Categorical(df_final['Results'], categories=custom_sort_order, ordered=True)
df_final = df_final.sort_values(by=['Location', 'Results']).reset_index(drop=True)
df_final
```
